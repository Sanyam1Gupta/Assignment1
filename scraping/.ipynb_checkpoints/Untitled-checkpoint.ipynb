```python
from bs4 import BeautifulSoup as soup
```

```python
from urllib.request import Request, urlopen
url="https://www.zomato.com/bangalore/south-bangalore-restaurants"
req = Request(url, headers={'User-Agent': 'XYZ/3.0'})

web_byte = urlopen(req).read()

webpage = web_byte.decode('utf-8')
```

```python
page_soup = soup(webpage, "html.parser")
```

```python
page_soup
```

```python
containers = page_soup.findAll("div", {"class": ["col-s-12"]})
```

```python
containers
```

```python
containers[2]
```

```python
rest_type=containers[1].findAll("a",{"class":"zdark"})
rest_type[0].text
```

```python
rest_name=containers[2].findAll("a",{"class":"result-title"})
rest_name=rest_name[0].text
jj=rest_name.split('\n')
rest_name=jj[0]
rest_name
```

```python
rest_area=containers[0].b
rest_area=rest_area.find(text=True)
rest_area
```

```python
k=1
rest_list=[]
for container in containers:
    
    rest_type=container.findAll("a",{"class":"zdark"})
    rest_type=rest_type[0].text
    rest_name=container.findAll("a",{"class":"result-title"})
    rest_name=rest_name[0].text
    jj=rest_name.split('\n')
    rest_name=jj[0]
    rest_area=container.b
    rest_area=rest_area.find(text=True)
    rest_area
    dict1 = {'restaurant_id': k, 'name': rest_name,'area': rest_area, 'restaurant_type': rest_type}
    rest_list.append(dict1)
    k=k+1
```

```python
rest_list
```

```python
url="https://www.zomato.com/bangalore/south-bangalore-restaurants?page=2"
req = Request(url, headers={'User-Agent': 'XYZ/3.0'})

web_byte = urlopen(req).read()

webpage = web_byte.decode('utf-8')
page_soup = soup(webpage, "html.parser")
containers = page_soup.findAll("div", {"class": ["col-s-12"]})
for container in containers:
    rest_type=container.findAll("a",{"class":"zdark"})
    if(len(rest_type)>0):
         rest_type=rest_type[0].text
    else:
         rest_type=""
    rest_name=container.findAll("a",{"class":"result-title"})
    rest_name=rest_name[0].text
    jj=rest_name.split('\n')
    rest_name=jj[0]
    rest_area=container.b
    rest_area=rest_area.find(text=True)
    rest_area
    dict1 = {'restaurant_id': k, 'name': rest_name,'area': rest_area, 'restaurant_type': rest_type}
    rest_list.append(dict1)
    k=k+1
```

```python
url="https://www.zomato.com/bangalore/south-bangalore-restaurants?page=6"
req = Request(url, headers={'User-Agent': 'XYZ/3.0'})

web_byte = urlopen(req).read()

webpage = web_byte.decode('utf-8')
page_soup = soup(webpage, "html.parser")
containers = page_soup.findAll("div", {"class": ["col-s-12"]})
for container in containers:
    if(k>80):
        break
    rest_type=container.findAll("a",{"class":"zdark"})
    if(len(rest_type)>0):
         rest_type=rest_type[0].text
    else:
         rest_type=""
    rest_name=container.findAll("a",{"class":"result-title"})
    rest_name=rest_name[0].text
    jj=rest_name.split('\n')
    rest_name=jj[0]
    rest_area=container.b
    rest_area=rest_area.find(text=True)
    rest_area
    dict1 = {'restaurant_id': k, 'name': rest_name,'area': rest_area, 'restaurant_type': rest_type}
    rest_list.append(dict1)
    k=k+1
```

```python
import json
with open('result.json', 'w') as fp:
    json.dump(rest_list, fp)
```

```python
import csv
```

```python
with open('result.csv', 'w', newline="") as csv_file:  
    writer = csv.writer(csv_file)
    for tt in rest_list:
       for key, value in tt.items():
           writer.writerow([key, value])
```
